```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
data = pd.read_csv("housing_data.csv")

# Display first few rows
print(data.head())

# Set plot style and color palette
sns.set(style="whitegrid", palette="husl")

# Create pair plot
sns.pairplot(data, vars=["Price", "Area", "Rooms"], diag_kind="kde", hue="City")

plt.suptitle("Relationships Among House Price, Area, and Rooms", y=1.02)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load Titanic dataset
titanic = sns.load_dataset("titanic")

# Drop missing values
titanic_clean = titanic.dropna(subset=["age", "fare", "pclass", "survived"])

# Set custom style
sns.set(style="darkgrid", palette="cool")

# Create pair plot
sns.pairplot(titanic_clean, vars=["age", "fare", "pclass"], hue="survived", diag_kind="kde")

plt.suptitle("Titanic Passenger Patterns: Survival by Age, Fare, and Class", y=1.02)
plt.show()
```
