```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv('population.csv')

# Sort and take top 5 countries
top5 = df.sort_values(by='Population', ascending=False).head(5)

# Convert population to billions for readability
top5['Population_Billions'] = top5['Population'] / 1_000_000_000

# Plot bar chart
plt.figure(figsize=(10,6))
plt.bar(top5['Country'], top5['Population_Billions'])

plt.title("Top 5 Most Populated Countries")
plt.xlabel("Country")
plt.ylabel("Population (in Billions)")

plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("movies_genre.csv")

# Plot bar chart
plt.figure(figsize=(12,6))
plt.bar(df['Genre'], df['Count'])

plt.title("Number of Movies Produced by Each Genre")
plt.xlabel("Genre")
plt.ylabel("Movie Count")

# Rotate labels for readability
plt.xticks(rotation=45)

plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
```
